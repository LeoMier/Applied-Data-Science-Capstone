### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
#First, we must search the name of the table

query_0 = """
SELECT name FROM sqlite_master WHERE type='table';
"""

cur.execute(query_0)
table_names = cur.fetchall()
print(table_names)

#Then, we can search for the column names
'''
query_1 ="""
PRAGMA table_info(SPACEXTBL);
"""

cur.execute(query_1)
column_names = cur.fetchall()
print(column_names)

#We print the full table

query_2 = """
SELECT *
FROM SPACEXTBL
"""
cur.execute(query_2)
table = cur.fetchall()
for row in table:
    print(row)

#we select the 'Lauch_Site' column
'''
query_3 = """
SELECT DISTINCT "Launch_Site"
FROM SPACEXTBL
"""
cur.execute(query_3)
lauch_site = cur.fetchall()
for row in lauch_site:
    print(row)
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
query_3 = """
SELECT *
FROM SPACEXTBL
WHERE "Launch_Site" LIKE 'CCA%'
LIMIT 5;
"""

cur.execute(query_3)
records = cur.fetchall()
for row in records:
    print(row)
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
query_5 = """
SELECT SUM("PAYLOAD_MASS__KG_")
FROM SPACEXTBL
WHERE  "Customer" ='NASA (CRS)';
"""

cur.execute(query_5)
print(cur.fetchall())
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
query_6 = """
SELECT AVG("PAYLOAD_MASS__KG_")
FROM SPACEXTBL
WHERE "Booster_Version" = 'F9 v1.1';
"""

cur.execute(query_6)
print(cur.fetchall())
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
query_7 = """
SELECT MIN("Date")
FROM SPACEXTBL
WHERE "Landing_Outcome" = "Success (ground pad)";
"""

cur.execute(query_7)
print(cur.fetchall())
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
query_8 = """
SELECT DISTINCT "Booster_Version"
FROM SPACEXTBL
WHERE "Landing_Outcome" = "Success (drone ship)" AND "PAYLOAD_MASS__KG_" BETWEEN 4000 AND 6000;
"""

cur.execute(query_8)
print(cur.fetchall())
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
query_9 = """
SELECT COUNT(*)
FROM SPACEXTBL
WHERE "Mission_Outcome" = "Success"
"""

query_10 = """
SELECT COUNT(*)
FROM SPACEXTBL
WHERE "Mission_Outcome" <> "Success"
"""

query_10 = """
SELECT COUNT(*)
FROM SPACEXTBL
WHERE "Mission_Outcome" <> "Success"
"""

cur.execute(query_9)
print(cur.fetchall())

cur.execute(query_10)
print(cur.fetchall())
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
query_11 = """
SELECT "Booster_Version", "PAYLOAD_MASS__KG_"
FROM (SELECT * FROM SPACEXTBL ORDER BY "PAYLOAD_MASS__KG_" DESC LIMIT 12);
"""

cur.execute(query_11)
data = cur.fetchall()

for row in data:
    print(row)
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 4, 2) as month to get the months and substr(Date,7,4)='2015' for year.**


```python

query_12 = """
ALTER TABLE SPACEXTBL ADD COLUMN month_name TEXT;
"""
query_13 = """
UPDATE SPACEXTBL
SET month_name = SUBSTR("Date", 4, 2);
"""

query_14 = """
SELECT *
FROM SPACEXTBL;
"""

query_16 = r"""
SELECT "month_name", "Landing_Outcome", "Booster_Version", "Launch_Site"
FROM (SELECT * FROM SPACEXTBL WHERE "Date" BETWEEN "01/01/2015" AND "31/12/2015")
WHERE "Landing_Outcome" LIKE '%drone ship%';
"""

cur.execute(query_12)
cur.execute(query_13)
cur.execute(query_16)
table = cur.fetchall()

for row in table:
    print(row)
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
query_15 = """
SELECT "Landing_Outcome" ,COUNT(*)
FROM (SELECT * FROM SPACEXTBL WHERE "DATE" BETWEEN "04/06/2010" AND "20/03/2017")
GROUP BY "Landing_Outcome"
ORDER BY COUNT(*) DESC;
"""

cur.execute(query_15)
rank = cur.fetchall()

for row in rank:
    print(row)
```
